```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
plt.style.use('seaborn')
```

```python
def display_csv_and_plot(csv_path='custom_results.csv', plot_path='custom_plot.png'):
    if not os.path.exists(csv_path):
        print(f"Файл {csv_path} не найден!")
        return
    if not os.path.exists(plot_path):
        print(f"Файл {plot_path} не найден!")
        return

    print("="*50)
    print(f"ДАННЫЕ ИЗ ФАЙЛА: {csv_path}")
    print("="*50)

    df = pd.read_csv(csv_path)

    print(df)

    print("\n" + "="*50)
    print(f"ГРАФИК ИЗ ФАЙЛА: {plot_path}")
    print("="*50)

    plt.figure(figsize=(12, 6))
    img = plt.imread(plot_path)
    plt.imshow(img)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

    print("\n" + "="*50)
    print("АНАЛИТИКА ДАННЫХ")
    print("="*50)
    print(f"Всего тестов: {len(df)}")
    print(f"Среднее время: {df['Multiplication Time (s)'].mean():.2f} с")
    print(f"Максимальное время: {df['Multiplication Time (s)'].max():.6f} c (размер {df.loc[df['Multiplication Time (s)'].idxmax(), 'Dimension']})")
    print(f"Минимальное время: {df['Multiplication Time (s)'].min():.6f} c (размер {df.loc[df['Multiplication Time (s)'].idxmin(), 'Dimension']})")

```

```python
display_csv_and_plot()
```
